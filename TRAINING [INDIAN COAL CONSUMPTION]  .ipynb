```python
coal_consumed = [10, 15, 8, 12, 9, 25, 6, 18, 11, 7]

plt.boxplot(coal_consumed)
plt.xlabel('Coal Consumed')
plt.ylabel('Amount')
plt.title('Distribution of Coal Consumed')
plt.show()
```

```python
coal_consumption = {'OTHER': 0.131, 'POWER (UTILITY)': 48.542, 'POWER (CAPTIVE)': 3.825, 'PULP & PAPER': 0.095, 'FERTILISERS': 1.111, 'CEMENT': 0.045, 'STEEL (BOILERS)': 0.121}

pie_chart_slice_labels = coal_consumption.keys()

pie_chart_slice_values = coal_consumption.values()
plt.pie(pie_chart_slice_values, labels=pie_chart_slice_labels)
plt.title("Pie Chart of Coal Consumption by Sector")
plt.show()
```
